# K Means Clustering


```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
df.head()
```

```python
plt.scatter(df['Age'],df['Income($)'])
```

```python
kmod=KMeans(n_clusters = 3)
kmod
```

```python
y_predicted=kmod.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df0=df[df.cluster == 0]
df1=df[df.cluster == 1]
df2=df[df.cluster == 2]
plt.scatter(df0.Age,df0['Income($)'],color='skyblue')

plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='pink')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
scalar = MinMaxScaler()
scalar.fit(df[['Income($)']])
df['Income($)'] = scalar.transform(df[['Income($)']])
```

```python
df
```

```python
scalar.fit(df[['Age']])
df['Age'] = scalar.transform(df[['Age']])
df
```

```python
y_predicted = kmod.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['Cluster'] = y_predicted 
df.drop(['cluster'],axis = 'columns', inplace = True)
df
```

```python
kmod.cluster_centers_
```

```python
df0=df[df.Cluster == 0]
df1=df[df.Cluster == 1]
df2=df[df.Cluster == 2]

plt.scatter(df0.Age,df0['Income($)'],color='skyblue')
plt.scatter(df1.Age,df1['Income($)'],color='green')
plt.scatter(df2.Age,df2['Income($)'],color='pink')
plt.scatter(kmod.cluster_centers_[:,0] , kmod.cluster_centers_[:,1],color = 'Red')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
# Excersice
```

```python
k_rng = range(1,15)
sse = []
for k in k_rng:
    kmod = KMeans(n_clusters = k )
    kmod.fit(df[['Age','Income($)']])
    sse.append(kmod.inertia_)
```

```python

sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
